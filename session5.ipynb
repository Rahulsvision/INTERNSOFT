```
def is_int():
    try:
        gg = int(input("Please enter the input : "))
    except:
        print("Error:The input is not integer")
        gg = int(input("Plese Try again : "))
    finally:
        print("Finally Done")
    print(gg)
is_int()
```

```
def is_float():
  while True:
    try:
        gg = float(input("Please enter the input : "))
    except:
        print("Error:The input is not float")
        continue
    else:
        print("We have it")
        break
    finally:
        print("completed task")
    print(gg)
is_float()
```

```
import sqlite3
n = sqlite3.connect("database.db")
n.execute("drop table if exists student")
n.execute("create table student(srn int, name text, gender text)")
n.execute("insert into student(srn, name, gender) values(242,'Rahul hk','Male')")
n.execute("insert into student(srn,name,gender) values(245,'Rakesh M','Male')")
n.execute("insert into student(srn,name,gender) values(246,'Girly','Female')")
n.execute("insert into student(srn,name,gender) values(247,'Rashi','Female')")
n.execute("insert into student(srn,name,gender) values(248,'Rohan L','Male')")
n.commit()
```

```
end = n.execute("select * from student order by name")
for row in end:
    print(row)
```
