```
#methods
class home:
    def __init__(self,numb,place,dob):
        self.numb = numb
        self.place = place
        self.dob = dob
    def age(self,current_dob):
        return current_dob - self.dob
    def __str__(self):
        print('%d %s %d'%(self.place,self.dob))
clc = home(371,'msr',54)
print(clc.age(90))
```

```
#inheritance
class Area(home):
    def __init__(self, area_name, *args, **kwargs):
        super(Area, self).__init__(*args, **kwargs)
        self._area_name = area_name
        
chill = home(3, 'name', 33)
print(type(chill))
print(isinstance(chill, home))
print(isinstance(chill, object))
```
