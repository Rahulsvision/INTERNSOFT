```
#boolean indexing
import numpy as np
a = np.arange(25).reshape(5,5)
a[a > 12] = 5
a[a == 5] = -1
print(a)
print(a == -1)
```

```
#boolean with 2 tests
a = np.arange(25).reshape(5,5)
b = np.arange(25).reshape(5,5)
a[np.logical_or(a<1,a>16)] = 0.0
b[np.logical_and(a>9,a<19)] = 0.0
print(a)
print(b)
a > 4
```

```
#avoid loops
g = 20
h = 80
smin = tmin = 5.0
smax = tmax = 10.0
s = np.linspace(smin, smax, g, endpoint=True)
t = np.linspace(tmin, tmax, h, endpoint=False)
print(s.shape)
print(t.shape)
print(s)
t
```

```
#time elapsed
import time
import numpy as np
a = 10
b = 50
c = np.linspace(1.0, 3.0, a, endpoint=True)
d = np.linspace(0.0, 4.0, b, endpoint=False)
T0 = time.time()
f = np.zeros((a, b))
for i in range(a):
    for j in range(b):
        f[i,j] = np.sin(2.0*np.pi*c[i]*d[j])
T1 = time.time()
print("time elapsed: {} s".format(T1-T0))
print(c)
d
```

```
#using array
cd, dd = np.meshgrid(c, d, indexing="ij")
print(cd[:,4])
print(dd[5,:])
t = time.time()
g2 = np.sin(2.0*np.pi*cd*dd)
t0 = time.time()
print("time elapsed: {} s".format(t0-t))
```

```
#concantinate
a = np.array([0, 1, 6, 4])
b = np.array([1, 7, 5, 8])
c = np.array([5, 2, 3])
print(np.concatenate([a,b,c]))
g = np.array([[1, 2],
             [4, 5]])
np.concatenate([g,g],axis=1)
```

```
#splitting of array
a = [1, 4, 3, 8, 7, 6 ,2, 1]
a1, a2, a3 = np.split(a, [2,4])
print(a1, a2, a3)
b = np.arange(15).reshape(3,5)
left, right = np.hsplit(b, [3])
print(b)
print(left)
print(right)
```

```
#aggregate
x = np.arange(1, 10)
print(x)
print(np.add.reduce(x))
print(np.multiply.reduce(x))
print(np.add.accumulate(x))
print(np.multiply.accumulate(x))
```
