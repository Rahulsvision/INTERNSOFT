```
#functions:def,return
#example:even or odd number
def ch(num):
    if num % 2 == 0:
      return 'even'
    else:
      return 'odd'
```

```
#generate fibonacci series
def fibo(n):
    a = 5
    b = 4
    
    for i in range(n):
        yield a
        a,b = b,a+b        
```

```
 for output in fibo(13):
        print(output)        
```
