```
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.plot(np.random.normal(1,5,10),'d')
```

```
%matplotlib inline
plt.plot(np.random.normal(0,10,10),'c',np.random.normal(0,6,8),'s')
```

```
hola = pd.Series(np.random.normal(size = 10))
print(hola)
print(hola.cumsum(0).plot(grid = True))
plt.plot(hola,'s')
```

```
dude = pd.DataFrame({'normal':np.random.normal(0,10,size = 20),
                 'gamma':np.random.gamma(1, size = 20),
                 'poisson':np.random.poisson(2,size = 20)})
print(dude)
dude.cumsum(0).plot()
dude.cumsum(0).plot(secondary_y = 'poisson')
```

```
a,dip = plt.subplots(ncols= 3, nrows= 1, figsize=(20,3))
for i,var in enumerate(['normal','gamma','poisson']):
    dude[var].cumsum(0).plot(ax = dip[i], title = var)
dip[0].set_ylabel('codes')
```

```
a.savefig('figure.jpg')
from IPython.display import Image
print(a.canvas.get_supported_filetypes())
Image('figure.jpg')
```

```
a1 = plt.figure()
dip1 = plt.axes()
no = np.linspace(0,50,500)
print(no)
dip1.plot(no,np.sin(no))
plt.plot(no,np.cos(no))
```

```
wow = np.linspace(1,45,900)
plt.plot(wow,np.cos(wow-2),color = 'red')
plt.plot(wow,np.cos(wow-6),color = 'b')
plt.plot(wow,np.sin(wow-1),linestyle = '--')
plt.plot(wow,np.sin(wow-3),linestyle = ':')
```

```
plt.plot(wow,np.tan(wow),color = 'g',linestyle = '--')
plt.xlim(2,15)
plt.ylim(-15,15)
```

```
plt.plot(wow,np.tan(wow),linestyle =":",color='y')
plt.title("tan wave")
plt.xlabel('x axis')
plt.ylabel('y axis')
```
