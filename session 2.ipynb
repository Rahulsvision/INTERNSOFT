```
#create,print,using double quotes in a string
s = 'hey'
print(s)
s = "I'll be fine here"
s
```

```
#string basics and indexing
r = "I'm attending classes"
print(len(r))
print(r[5])
print(r[4:])
print(r[::3])
```

```
#string properties
x = 'what'
x = x + ' are you doing '
print(x)
x = x*2
print(x)
```

```
#builtin methods 
f = 'can everyone hear'
print(f.split('e'))
print(f.count('r'))
print(f.find('o'))
print(f.upper())
```

```
#format and check methods
a = ' Hey\twhat\tdo\tyou\tthink '.expandtabs()
print(a.center(70,'x'))
print(a.isalnum())
print(a.istitle())
print(a.islower())
print(a)
print(a.endswith('k'))
```

```
#lists and indexing
h = ['hey',25,1.2,'e']
print(h)
print(h[2])
print(h[:4])
print(h[::-1])
```

```
#list cancatination
k = ['hello',1,23,2.4,56]
print(k)
k = k + ['done',45,73]
print(k)
k = k*3
print(k)
```

```
#list methods
y = ['ad','food','hi',2,2.5,33,]
print(y)
y.append('god')
print(y)
y.pop(4)
print(y)
y.reverse()
print(y)
```

```
#nested list
l1 = [3,5,6]
l2 = [2,7,8]
l = [l1,l2]
print(l)
new = [row[0] for row in l]
print(new)
new = [row[2] for row in l]
print(new)
```

```
#list append,insert,extend
was = [12,2,25,5]
was.append([6,63])
print(was)
was.extend([58,99,48])
print(was)
was.insert(3,'hello')
was
```

```
#list sort,remove,reverse
was.remove('hello')
print(was)
was.reverse()
print(was)
was.remove([6,63])
print(was)
was.sort()
print(was)
```
