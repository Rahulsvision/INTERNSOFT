```
#create an array
import numpy as np
a = np.arange(25)
b = np.arange(30)
c = np.arange(27).reshape(3,9)
print(a)
print(b)
c
```

```
#meta data of an array
print(c.ndim)
print(c.shape)
print(c.size)
print(c.itemsize)
c.dtype
```

```
#array from list
d = np.array([5,10,15,20,25,30])
print(type(d))
print(d.size)
print(d.ndim)
print(d.itemsize)
d.shape
```

```
#eye,linespace,logspace functions
x = np.eye(5)
dtype = np.float64
y = np.linspace(1,5,10,endpoint=True)
z = np.logspace(1,5,10,endpoint=False,base=10)
print(dtype)
print(x)
print(y)
z
```

```
#lambda function,array operation
v = np.fromfunction(lambda a,b:a==b,(4,4),dtype=float)
u = np.arange(15).reshape(3,5)
print(v)
print(u)
print(u+u)
print(u*2)
print(u*u)
#matrix multiplication
z = u.transpose()
print(u@z)
```

```
#axis operation and extreme functions
e = np.array([1,2,4,5,3,8,9,7,5,6,11,12]).reshape(3,4)
print(e)
print(e.sum(axis=0))
print(e.sum(axis=1))
print(e.min(),e.max())
print(e.sum())
```

```
#universal functions
s = e*np.pi/6.0
print(s)
t = np.tan(s)
print(t)
q = t-s
q
```

```
#multidimensional array
ma = np.arange(50).reshape(5,10)
print(ma)
print(ma[3,2])
print(ma[3:])
print(ma[:,1:3])
print(ma[1:5,:])
m = ma.flatten()
print(m)
for i in m:
    print(i)
```

```
#copying arrays, referencing
rr = np.arange(6)
print(rr)
sky = rr
print(sky)
print(rr is sky)
sky.shape = (3,2)
print(sky)
rr.shape
```

```
#shallow copying
ra = np.arange(15).reshape(3,5)
print(ra)
hu= ra[:]
print(hu)
print(hu is ra)
hu.shape = (5,3)
print(hu)
ra.shape
```

```
#Checking arrays
print(hu is rr)
print(hu.base is rr)
print(hu.flags.owndata)
print(rr.flags.owndata)
```

```
#Deep copy
ar = rr.copy()
ar[:] = 0.0
print(rr)
ar
```
