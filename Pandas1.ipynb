```
#Panda version
import pandas
print(pandas.__version__)
```

```
#series in pandas
import pandas as pd
hola = pd.Series([23, 51, 42, 67, 33])
print(hola)
print(hola.index)
hola.values
```

```
#examples by assigning index names
bike = pd.Series([100, 23, 25, 88, 77],
        index=['Dominar','Activa','Splendor','Royal Enfield','KTM'])
print(bike)
print(bike.index)
print(bike[4])
print(bike['Royal Enfield']) 
```

```
#naming of the values and indexes
bike.name = "comfy"
bike.index.name = "store"
bike
```

```
#numpy operations in pandas series
import numpy as np
print(np.sin(bike))
print("")
print(np.log(bike))
print("")
print(np.tan(bike))
print("")
print(np.pi*bike)
print("")
```

```
#creating series from dictionary
playlist = {1:'Mere sapno ki rani',2:'Chala jata hoon',3:'Pal pal dil ke pas',4:'Musafir hoon yaron',5:'Roop tera Mastana'}
print(playlist)
print("")
playlist1 = pd.Series(playlist)
print(playlist1)
print("")
playlist1.index.name = "name"
playlist1.name = "play"
print(playlist1)
```

```
#dataframes
Awards = {"Singer":["Kishore Kumar","Kumar sanu","Asha","Mohamad Rafi"],
              "songs":[180,45,256,250],
              "awards":[18,10,17,36]}
print(Awards)
print("")
award = pd.DataFrame(Awards)
print(award)
print("")
award["Total"] = (award.songs + award.awards)/2
print(award)
print("")
award["Gender"] = pd.Series(["M","M","F","M"])
award
```

```
#conversion
print(award.to_dict())
print('''
       ''')
pd.DataFrame(award)
award
```

```
#dataframes from a list of dicts
squares_and_cubes = [{"x":i**3,"y":i**2}for i in range(10)]
print(squares_and_cubes)
print('''

      ''')
pd.DataFrame(squares_and_cubes)
```
