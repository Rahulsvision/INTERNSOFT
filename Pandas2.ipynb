```
#ufuncs index preservation
done = np.random.RandomState(25)
day = pd.Series(done.randint(1, 15, 6))
f = pd.DataFrame(done.randint(1, 20, (4, 6)),
                  columns=['can', 'call', 'cat', 'curry','command','calicut'])
print(day)
f
```

```
print(np.sin(day * np.pi / 2))
print('')
print(np.cosh(day * np.pi))
print('')
print(np.tanh(day * np.pi /4))
np.sinh(f*np.pi/2)
```

```
goals = pd.Series({'messi':130,'ronaldo':129,
                    'neymar':115,'zlatan':58},
                     name='srortsmen')
cups = pd.Series( {'neymar':70,'zlatan':32,
                    'messi':93,'ronaldo':95},name = 'average')
print(goals)
print('')
print(cups)
print('')
print(goals*cups)
print('')
print(goals.index | cups.index)
```

```
studs = {"Name":["appu","nody","bean","jack","sky","jaggu"],
              "ia1":[18,30,17,15,29,25],
              "ia2":[17,25,25,10,30,28]}
stud = pd.DataFrame(studs)
stud["class"] = pd.Series(['f','c','a','a','c','f'])
stud["avg"] = (stud.ia1 + stud.ia2)/2
stud["gender"] = pd.Series(["M","M","M","M","F","M"])
stud2 = pd.DataFrame({"class":['c','a','f','c','a'],'ia3' : [19,22,30,10,23]})
print(stud)
print('')
print(stud2)
print('')
stud.merge(stud2)
```
